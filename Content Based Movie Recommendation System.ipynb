```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
movies.head()
```

```python
credits.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
movies.info()
```

```python
# Drop the unwanted columns 

movies = movies[['id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L1 = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L1.append(i['name'])
    return L1
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies_new = movies[['id', 'title', 'tags']]
```

```python
movies_new.sample(5)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text:
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
movies_new['tags'] = movies_new['tags'].apply(stem)
```

```python
movies_new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies_new['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    index = movies_new[movies_new['title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
